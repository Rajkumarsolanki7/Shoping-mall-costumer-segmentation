```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Distribution of Age
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Distribution of Annual Income
plt.figure(figsize=(10, 6))
sns.histplot(df['Annual Income'], bins=20, kde=True)
plt.title('Annual Income Distribution')
plt.xlabel('Annual Income')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Distribution of Spending Score
plt.figure(figsize=(10, 6))
sns.histplot(df['Spending Score'], bins=20, kde=True)
plt.title('Spending Score Distribution')
plt.xlabel('Spending Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Boxplot of Age by Gender
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='Age', data=df)
plt.title('Age Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

```python
# Boxplot of Annual Income by Gender
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='Annual Income', data=df)
plt.title('Annual Income Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Annual Income')
plt.grid(True)
plt.show()
```

```python
# Boxplot of Spending Score by Gender
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='Spending Score', data=df)
plt.title('Spending Score Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Spending Score')
plt.grid(True)
plt.show()
```
